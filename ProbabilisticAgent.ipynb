```python
class probabilistic_agent():
    def __init__(self, total_dice, roll, last_bid = None):
        self.last = last_bid
        self.total_dice = total_dice
        self.roll = roll

    def make_bid(self):
        
        #Not the first bid!!
        if self.last != None:
            probability_cut = self.total_dice / 6 #multiply by 2 if not ace
            player_roll_mode = stats.mode(self.roll) #Mode of players roll
            
            #Call of aces!
            if self.last[1] == 1: #if previous bid was a bid of aces
                if self.last[0] > probability_cut: #if last bid is above avg
                    bid = 'Dudo'
                else: #if still reasonable quantities
                    if player_roll_mode[0] == 1: #and current_players roll mode is aces too
                        bid = [int((self.last[0])+1), int(player_roll_mode[0])] #just bid one more ace
                    else: #players mode isn't an ace
                        #bid has to be calling from aces to a normal number
                        bid = [int(((self.last[0])*2)+1), int(player_roll_mode[0])]
            else: #if previous bid was of something that wasn't aces!
                if self.last[0] > (probability_cut * 2): # *2 because we are counting with jokers too
                    bid = 'Dudo'
                else:
                    if player_roll_mode[0] == 1: #and current_players roll mode is aces
                        if self.last[0] % 2 == 0: #if its an even number
                            bid = [int(math.ceil((self.last[0])/2)), int(player_roll_mode[0])] #just bid one more ace
                        else:
                            round_up_quant = math.ceil(self.last[0]/2)
                            bid = [int(round_up_quant), int(player_roll_mode[0])]
                    else: #players mode isn't an ace
                        #bid has to be calling from aces to a normal number
                        bid = [int((self.last[0])+1), int(player_roll_mode[0])]
                        
                        
        #FIRST BID IN THE ROUND!
        else: 
            #It's not very risky to bluff at this stage of the game
            if self.total_dice > 15:
                bid_quant = rnd.randint(2, 3) #bid 2 or 3 for quantity at the start
            #if little dice in game, better to start low...
            else: 
                bid_quant = rnd.randint(1,2)
            
            #Analyse player's roll to bid
            player_roll_mode = stats.mode(self.roll)
                     
            if player_roll_mode[0] == 1: 
                #players can't start bidding on 1, so bid on random number
                random_number = int(rnd.randint(2,6))
                bid = [bid_quant, random_number] 
            else: # if the mode is any value other than aces and you're not bluffind, bid on mode:
                roll_mode_val = int(player_roll_mode[0]) 
                bid = [bid_quant, roll_mode_val]
    
        return bid
```
